```python
""" This Is Jobot, a cahtbot that will help you know more about jordan,
    
    it's a questions based chatbot that will start the conversation with you after you
    
    type a greeting word, you can ask him about specific places in jordan and you'll 
    
    find out about these places while you keep asking, if you want to know informations 
    
    about a specific place or jordan in general just type [about] then the place that 
    
    you want to know informations about it. """


import random
import csv
import re

greeting_keywords = ["hello", "hi", "what's up","hi there","is anyone there?","hey"]
greeting_responses = ["'sup bro", "hey", "hello","Hi there","hey"]

goodbye_words = ["goodbye","see you later","bye"]
goodbye_responses = ["goodbye","see you later", "take care","bye bye"]


def check_greeting(sentence):
    """this function checks if the user is greeting JObot """
    
    
    print(random.choice(greeting_responses)) 
    while True:
        sentence = input("<<<")
        if sentence in greeting_keywords:
            print("how can i help you?")
        elif sentence in goodbye_words:
            end_chat(sentence)
            break
        else:
            questions(sentence)
                
                             
def questions(sentence):
    """in this function the user can ask questions about jordan and the chatbot will answer."""
    
    
    with open("dataset12.csv",'r') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            x = row["questions"]
            if sentence in x:
                if re.findall(r".+?visit\sjordan|what's.+?name.?|how\s\w+?\syou.?|what's.+?age.?|what.+do?|about\sjordan|can.+?question.|what.+places.+in\sjordan.|.*old.*|.*about\sdowntown.*|is.+historical\splaces.|.*places\sin\sjordan|.*about\spetra.*|.+?see.+?(nature)|.*about\sjerash.*|.*(activities).*(jerash)|.*about\sum\sqais.*|.*(activities).*(um\sqais)|.*(adventures)|.*about\saqaba.*|.*(activities).*(aqaba)|.*about\swadi\srum.*|.*(prices).|.*people?|.*(activities).*(wadi\srum).*|tell.+?castles.|is.+?streets.|.*prices.*|how's.*?people.*?jordan|is.*ancient\sbuildings.?|tell.+?\scastels.|.*cities\sof\sjordan|.*?culture\sin\sjordan.?|about\sirbid|about\smadaba|about\skarak|about\shammamat\sma'in.*|about\sdead\ssea.?|thank you|.*meuseums\sin\sjordan.?|.*?(inexpensive).*?|places\sto\shang\sout|where.+?view\sof\samman.?|.*climate|what\sabout\stransportation.?|about\samman.?|help",sentence):
                    print(row["answers"])     
                
                elif sentence in goodbye_words:
                    end_chat(sentence)
                    break
                else:
                    print("i don't understand!, ask again or type help ")
                    break
                    
                    
    csvfile.close()               
                    
                    

def end_chat(sentence):
    '''This function will end the conversation when you type a one of the goodbye words. '''
    
    
    print(random.choice(goodbye_responses))                            
    


sentence = input('...')
while True:
    if sentence.lower() in greeting_keywords:
        check_greeting(sentence)
    elif sentence.lower() in goodbye_words:
            end_chat(sentence)
    else:
        questions(sentence)
            
    break        

            
            
    
```
